## 2. Build a model by subclassing `nn.Module` that incorporates non-linear activation functions and is capable of fitting the data you created in 1.

## 4. Create a training and testing loop to fit the model you created in 2 to the data you created in 1.
  * Do a forward pass of the model to see what's coming out in the form of logits, prediction probabilities and labels.
  * To measure model accuray, you can create your own accuracy function or use the accuracy function in [TorchMetrics](https://torchmetrics.readthedocs.io/en/latest/).
  * Train the model for long enough for it to reach over 96% accuracy.
  * The training loop should output progress every 10 epochs of the model's training and test set loss and accuracy.

```python
def accuracy(preds, labels):
    """
    Compute the accuracy, given the predictions and labels.
    Args:
    preds (torch.Tensor): Predicted outputs.
    labels (torch.Tensor): Ground truth labels.
    Returns:
    float: Accuracy in percentage.
    """
    predicted = (torch.sigmoid(preds) > 0.5).float()  # Apply sigmoid and threshold at 0.5
    correct = (predicted == labels).float().sum()
    return correct / len(labels)

# Step 5: Training and testing loops
epochs = 200
train_losses = []
test_losses = []
train_accuracies = []
test_accuracies = []

for epoch in range(epochs):
    # Training phase
    model.train()  # Set the model to training mode
    optimizer.zero_grad()  # Zero the gradients

    # Forward pass
    train_outputs = model(train_features_tensor)
    train_loss = criterion(train_outputs, train_labels_tensor.unsqueeze(1))

    # Backward pass and optimization
    train_loss.backward()
    optimizer.step()

    # Calculate training accuracy
    train_acc = accuracy(train_outputs, train_labels_tensor)

    # Append train loss and accuracy
    train_losses.append(train_loss.item())
    train_accuracies.append(train_acc.item())

    # Testing phase
    model.eval()  # Set the model to evaluation mode
    with torch.no_grad():
        test_outputs = model(test_features_tensor)
        test_loss = criterion(test_outputs, test_labels_tensor.unsqueeze(1))
        test_acc = accuracy(test_outputs, test_labels_tensor)

        # Append test loss and accuracy
        test_losses.append(test_loss.item())
        test_accuracies.append(test_acc.item())

    # Output progress every 10 epochs
    if (epoch + 1) % 10 == 0:
        print(f'Epoch [{epoch + 1}/{epochs}]')
        print(f'Training Loss: {train_loss.item():.4f}, Training Accuracy: {train_acc:.4f}')
        print(f'Test Loss: {test_loss.item():.4f}, Test Accuracy: {test_acc:.4f}')
        print('-' * 40)

    # Stop training if test accuracy exceeds 96%
    if test_acc >= 0.99:
        print(f'Reached 99% accuracy on the test set at epoch {epoch + 1}')
        break
```

## 5. Make predictions with your trained model and plot them using the `plot_decision_boundary()` function created in this notebook.

```python
# Plot the model predictions
import numpy as np

def plot_decision_boundary(model, X, y):

    # Put everything to CPU (works better with NumPy + Matplotlib)
    model.to("cpu")
    X, y = X.to("cpu"), y.to("cpu")

    # Source - https://madewithml.com/courses/foundations/neural-networks/
    # (with modifications)
    x_min, x_max = X[:, 0].min() - 0.1, X[:, 0].max() + 0.1
    y_min, y_max = X[:, 1].min() - 0.1, X[:, 1].max() + 0.1
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 101),
                         np.linspace(y_min, y_max, 101))

    # Make features
    X_to_pred_on = torch.from_numpy(np.column_stack((xx.ravel(), yy.ravel()))).float()

    # Make predictions
    model.eval()
    with torch.inference_mode():
        y_logits = model(X_to_pred_on)

    # Test for multi-class or binary and adjust logits to prediction labels
    if len(torch.unique(y)) > 2:
        y_pred = torch.softmax(y_logits, dim=1).argmax(dim=1) # mutli-class
    else:
        y_pred = torch.round(torch.sigmoid(y_logits)) # binary

    # Reshape preds and plot
    y_pred = y_pred.reshape(xx.shape).detach().numpy()
    plt.contourf(xx, yy, y_pred, cmap=plt.cm.RdYlBu, alpha=0.7)
    plt.scatter(X[:, 0], X[:, 1], c=y, s=40, cmap=plt.cm.RdYlBu)
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
```
